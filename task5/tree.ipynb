```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
dataset = pd.DataFrame(
    {
        'Age': ['young', 'young', 'medium', 'old', 'old', 'old', 'medium', 'young', 'young', 'old', 'young', 'medium', 'medium'], 
        'Incoming': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high'], 
        'Student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes','no', 'yes', 'yes', 'yes', 'no', 'yes'],
        'Credit Rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair'], 
        'Buying': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes']
    }
)
print(dataset)

age_map = {'young': 0, 'medium': 1, 'old': 2}
incoming_map = {'low': 0, 'medium': 1, 'high': 2}
student_map = {'no': 0, 'yes': 1}
credit_map = {'fair': 0, 'excellent': 1}
buying_map = {'no': 0, 'yes': 1}

dataset['Age'] = dataset['Age'].map(age_map)
dataset['Incoming'] = dataset['Incoming'].map(incoming_map)
dataset['Student'] = dataset['Student'].map(student_map)
dataset['Credit Rating'] =dataset['Credit Rating'].map(credit_map)
# dataset['Buying'] = dataset['Buying'].map(buying_map)
X = dataset[['Age', 'Incoming', 'Student', 'Credit Rating']]
Y = dataset[['Buying']]
cart_tree = DecisionTreeClassifier(criterion='gini').fit(X, Y)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(
    cart_tree, 
    feature_names=['Age', 'Incoming', 'Student', 'Credit Rating'],  
    class_names=['no', 'yes'],
    filled=True
)
```
